###### Formatting string using format() method

Format() method was introduced with Python3 for handling complex string formatting more efficiently. Formatters work by putting in one or more replacement fields and placeholders defined by a pair of curly braces { } into a string and calling the str.format(). The value we wish to put into the placeholders and concatenate with the string passed as parameters into the format function. 

````python
x=10
y=3
z=x/y

print('When we divide {} by {} we get {}'.format(x,y,z))
````

####### Left right and centre alighnment
Value can also be left, right, or centre aligned using python format() method.

For displaying a number with left alignment, we use the “:<n” symbol inside the placeholder in the format() method.

For displaying a number with the right alignment, we use the “:>n” symbol inside the placeholder.

To display a number with centre alignment, we use the “:^n” symbol inside the placeholder.  

Here "n" is the total length of the required 

````python
a='hello'
b='World'

print('hi,{:>10} {}'.format(a,b))
````

````python
a='hello'
b='World'

print('hi,{:>10} {:<5}'.format(a,b))
````

````python
a='hello'
b='World'

print('hi,{:<10}tictoc{:>25}.'.format(a,b))
````

````python
a='hello'
b='World'

print('hi,{:^10}tictoc{:^10}.'.format(a,b))
````

####### Padding Strings using the format() method

We use the same syntax we used for alignment operation to perform padding on the strings. Alignment is a specific type of padding in which the padding character is the space character.

To pad strings with other characters, we can use the syntax as defined below. Here we assume that we have to pad the strings using the * character.

To pad a string on the left side with , we use the “:<n” symbol inside the placeholder. 

To pad a string on the right side, we use the “:*>n” symbol inside the placeholder. 

To pad a string keeping the input string in the center, we use the “:*^n” symbol inside the placeholder.

Here "n" is the total length of the required output string.

````python
a='hello'
b='World'

print('hi,{} {:*<7}.'.format(a,b))
````

Here World is having 5 charecters plus 2 for * in total 7

````python
a='hello'
b='World'

print('hi,{} {:*>7}.'.format(a,b))
````

````python
a='hello'
b='World'

print('hi,{} {:*^7}.'.format(a,b))
````
